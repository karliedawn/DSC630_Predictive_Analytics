```python
# Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Import data
df = pd.read_csv('Medicalpremium_dataset.csv')
```

```python
# Look at data
df.head()
```

***Write a summary of your data and identify at least two questions to explore visually with your data.***

This dataset was found on Kaggle as a resource for data science students. The data contains factors related to the physical health and condition of an individual. The variables 'Age', 'Height', 'Weight', 'PremiumPrice' and 'NumberOfMajorSurgeries' are numeric variables, while 'Diabetes', 'BloodPressureProblems', 'AnyTransplants', 'AnyChronicDiseases', 'KnownAllergies', and 'HistoryOfCancerInFamily' are binary categorical variables. This data was provided by the individuals themselves and was given voluntarily. I would like to know if this data contains balanced classes for these features, and if any of these features are correlated with PremiumPrice.

***Create a histogram or bar graph from your data.***

```python
# histogram of height distribution
plt.title("Histogram of Height")
plt.hist(df['Height'], bins='auto', density=True)
plt.xlabel("Height (cm)")
plt.ylabel("Frequency")
plt.show()
```

***Create a boxplot from your data.***

```python
# boxplot of weight distribution
plt.boxplot(df['Weight'])
plt.title("Boxplot of Weight")
plt.ylabel("Weight (kg)")
plt.show()
```

***Create a bivariate plot from your data.***

```python
# Scatterplot of Age vs. Premium
plt.scatter(df['Age'], df['PremiumPrice'], alpha=0.5, color='purple')
plt.title(f"Scatterplot of Age vs. Premium Price", fontsize = 15)
plt.ylabel("Premium Price in $", fontsize = 8)
plt.xlabel(f'Age (Years)', fontsize = 8)
plt.show()
```

***Create any additional visualizations that will help to answer the question(s) you want to answer.***

```python
# correlation heat map
sns.heatmap(df.corr())
```

```python
# make histograms for each feature
for column in df:
    plt.title("Histogram of " + column)
    plt.hist(df[column], bins='auto', density=True)
    plt.show()
```

***Summarize your results and make a conclusion. Explain how you arrived at this conclusion and how your visualizations support your conclusion.***

In conclusion, we can see that the features 'Age' and 'NumberOfMajorSurgeries' correlate the most strongly with 'PremiumPrice' compared with the other features available in the dataset. They may be helpful in creating a model that predicts the premium price for an individual given their health history. This can be seen both on the correlation heatmap as well as it is visually detectable on the scatterplot of 'Age' vs. 'PremiumPrice'. 

Futhermore, we can see that the features of 'Age', 'Diabetes', 'BloodPressureProblems', and 'Height' had relatively balanced classes but 'AnyTransplants, 'AnyChronicDisease', 'Weight', 'HistoryOfCancerInFamily', and 'NumberOfMajorSurgeries' had unbalanced classes. This is just a matter of looking at the ratio of the heights of the bars on each of the feature histograms shown above.
